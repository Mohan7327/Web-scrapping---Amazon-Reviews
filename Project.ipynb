```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.amazon.in/product-reviews/B089MTJVLD/ref=acr_dp_hist_5?ie=UTF8&filterByStar=five_star&reviewerType=all_reviews&pageNumber=1'
```

```python
def get_soup(url):
  page = requests.get(url)
  soup = bs(page.content,'html.parser')
  return soup
```

```python
Scraped_reviews=[]
Scraped_body=[]
Scraped_Stars= []
```

```python
def get_reviews(soup):
  reviews = soup.find_all('a',class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
  try:
    for review in reviews:
       review = review.get_text().replace('\n',' ')
       review = review.strip()
       Scraped_reviews.append(review)
  except:
      pass
```

```python
def get_body(soup):
  body = soup.find_all('div',class_='a-row a-spacing-small review-data')
  try:
    for rev in body:
       rev = rev.get_text().replace('\n',' ')
       rev = rev.strip()
       Scraped_body.append(rev)
  except:
      pass
```

```python
def get_stars(soup):
  Stars = soup.find_all('i',class_='review-star-rating-view-point')
  try:
    for Star in Stars:
       Star = Star.get_text().replace('\n',' ')
       Star = Star.strip()
       Scraped_Stars.append(Star)
  except:
      pass
```

```python
for x in range(1,3000):
  soup = get_soup(f'https://www.amazon.in/product-reviews/B089MTJVLD/ref=acr_dp_hist_5?ie=UTF8&filterByStar=five_star&reviewerType=all_reviews&pageNumber={x}')
  get_body(soup)
  if not soup.find('li',class_='a-disabled a_last'):
      pass
  else:
      break
```
